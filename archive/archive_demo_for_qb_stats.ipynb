```python
import polars as pl
import nfl_data_py as nfl

SEASONS = [2025]
SEASON_TYPE = "REG"
MIN_DROPBACKS = 100

# Load PBP with nfl_data_py (returns pandas)
pdf = nfl.import_pbp_data(
    years=SEASONS,
    downcast=True,
    cache=False,
    alt_path=None,
    include_participation=False,
)

# Convert to Polars
df = pl.from_pandas(pdf)

# Filter season_type
if "season_type" in df.columns:
    df = df.filter(pl.col("season_type") == SEASON_TYPE)

df
```

```python
if "pass_attempt" not in df.columns:
    raise RuntimeError("Expected 'pass_attempt' column missing.")

pass_plays = df.filter(pl.col("pass_attempt") == 1)
pass_plays.head()
```

```python
#EPA and CPOE per QB -- all Qbs

group_cols = ["season", "posteam", "passer_player_name"]

qb_stats = (
    pass_plays
    .group_by(group_cols)
    .agg([
        pl.col("pass_attempt").sum().alias("dropbacks"),
        pl.col("epa").mean().alias("epa_per_pass"),
        pl.col("cpoe").mean().alias("cpoe"),   # ignores nulls automatically
    ])
)
qb_stats
```

```python
qb_filtered = (
    qb_stats
    .filter(pl.col("dropbacks") >= MIN_DROPBACKS)
    .sort("epa_per_pass", descending=True)
    .with_columns([
        pl.col("epa_per_pass")
        .rank(method="dense", descending=True)
        .cast(pl.Int64)
        .alias("epa_rank")
    ])
    .sort("epa_rank")
)

qb_filtered.head(50)
```

```python


# Choose metric to rank by: "epa_per_pass" or "cpoe"
#metric = "epa_per_pass"   # change to "cpoe" when you want to rank by CPOE
metric = "cpoe"


# Filter to QBs with enough volume
qb_filtered = qb_stats.filter(pl.col("dropbacks") >= MIN_DROPBACKS)

# Name the rank column based on the metric
rank_col = f"{metric}_rank"

# Compute rank (higher metric value = better rank, i.e. 1 is best)
qb_ranked = (
    qb_filtered
    .sort(metric, descending=True)
    .with_columns(
        pl.col(metric)
        .rank(method="dense", descending=True)
        .cast(pl.Int64)
        .alias(rank_col)
    )
    .sort(rank_col)
)

# Convert to pandas for nicer display (removes dtype row in output)
qb_ranked_pd = qb_ranked.to_pandas()

# Show the columns you care about
cols_to_show = [
    "season",
    "posteam",
    "passer_player_name",
    "dropbacks",
    "epa_per_pass",
    "cpoe",
    rank_col,
]

qb_ranked_pd[cols_to_show].head(32)
```
